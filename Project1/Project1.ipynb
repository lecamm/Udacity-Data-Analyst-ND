```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams.update({'font.size': 16})
```

```python
city=pd.read_csv('city.csv')
global_temp=pd.read_csv('global.csv')
df=city.merge(global_temp,on='year',how='inner',suffixes=('_city','_glob'))
#df.year=pd.to_datetime(df.year, format='%Y')
#df=df.set_index('year')
df.head()
```

```python
plt.figure(figsize=(14,8))
plt.plot(df.rolling(10).mean(),'.-')
plt.grid(linestyle=':')
plt.title('10-Year Moving Average')
plt.legend(['Dublin','Global'])
plt.xlabel('Time [years]')
plt.ylabel('Temperature [C]')
#plt.plot(df['avg_temp_city'],'b.')
#plt.plot(df['avg_temp_glob'],color='orange',marker='.')
plt.show()
```
